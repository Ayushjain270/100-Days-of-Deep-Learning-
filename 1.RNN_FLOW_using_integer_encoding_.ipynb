```python
# Our goal is to do the sentiment analysis ..
# is the sentence is positive then 1
# else 0



# 1. first we find the unique word in whole data
# 2. then we will do interger encoding /embeddings of the unique word
# 3 .do the the padding .. to make all the interger encoding  in the same length
```

```python
import numpy as np
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
# first we will do the tokenization ...
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer # if seperate word by word and make it small letter and do the indexing
tokenizer = Tokenizer(oov_token="<any_word>") # in the time of prediction if we give new word
#.. then the new word will be handle by oov_token ..
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index  # giving index to unique words
```

```python
tokenizer.word_counts # how many words repeats
```

```python
tokenizer.document_count # no. of sentences / rows
```

```python
sequences  = tokenizer.texts_to_sequences(docs)
sequences
```

```python
# now we will apply padding as the all the sequences has different length
```

```python
from keras.utils import pad_sequences
```

```python
sequence2 = pad_sequences(sequences,padding='post')
```

```python
sequence2
```

```python
# now we will apply on the actual data
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN , Embedding , Flatten
```

```python
(X_train , y_train) ,(X_test , y_test) = imdb.load_data()
```

```python
X_train[0] # already done interger encdoing by the keras
```

```python
print(len(X_train[0])) # we can see that the length are different so we have to do the padding
print(len(X_train[1]))
```

```python
X_train = pad_sequences(X_train,padding='post' , maxlen = 50)
X_test = pad_sequences(X_test,padding='post' ,maxlen = 50)
```

```python
X_train
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(SimpleRNN(32,input_shape=(2494,1) , return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
